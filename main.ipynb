```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray,(5,5),0)
imgCanny = cv2.Canny(blurred,75,200)
cv2.imshow("dd" ,imgCanny)
cv2.waitKey(0)
```

```python
contours, hierarchy = cv2.findContours(imgCanny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
# Iterate through each contour

for contour in contours:
    # Calculate the area of the contour
    area = cv2.contourArea(contour)
    # If the area is greater than the minimum area limit, draw the contour
    if area > 60000:
        cv2.drawContours(img, [contour], -1, (0, 255,0 ), 10)
cv2.imshow('Cnt',img)
cv2.waitKey(0)
# imageBlank=np.zeros_like(img)
# imgarray = ([img,imgGray,imgBlur,imgCanny],[imgContours,imageBlank,imageBlank,imageBlank])
# imgstacked = utlis.stackImages(imgarray,0.5)
# Get the bounding rectangles for each contour
rects = [cv2.boundingRect(contour) for contour in contours if cv2.contourArea(contour) > 60000]
```

```python


# Sort the rectangles from left to right based on their x-coordinate values
rects_sorted = sorted(rects, key=lambda x: x[0])

# Initialize a list to store parts of each rectangle
rect_parts = []

# Loop through each sorted rectangle and cut it into 30 parts
for rect in rects_sorted:
    # Calculate the width and height of each part
    part_width = rect[2]
    part_height = rect[3] // 30

    # Initialize a list to store parts of the current rectangle
    parts = []

    # Loop through each part and extract it from the image
    for i in range(30):
        part = imgCanny[rect[1]+i*part_height : rect[1]+(i+1)*part_height, rect[0]:rect[0]+part_width]
        
           
        # Divide the part into 4 subparts
        subpart_width = part_width // 4
        subparts = []
        for j in range(4):
            subpart = part[:, j*subpart_width : (j+1)*subpart_width]
            subparts.append(subpart)
      
        # Process each subpart to evaluate the OMR choices
        subpart_counts = []
        for subpart in subparts:
            # Count the number of non-zero pixels in the subpart
            pixel_count = cv2.countNonZero(subpart)
            
            # Append the pixel count to the subpart counts list
            subpart_counts.append(pixel_count)
        
        # Find the index of the subpart with the minimum pixel count
        min_idx = np.argmin(subpart_counts)
        
        # Append the index to the parts list
        parts.append(min_idx)
    
    # Append the parts of the current rectangle to the list of rectangle parts
    rect_parts.append(parts)

seq = np.array(rect_parts).reshape(150, 1)
seq
```

```python
# Set the random seed for reproducibility
rng = np.random.default_rng(seed=42)

# Generating model answer
model_answer = rng.integers(low=0, high=4, size=(150, 1))

model_answer
```

```python
# Count the number of correct values
num_correct = np.sum(seq == model_answer)
num_correct
```
